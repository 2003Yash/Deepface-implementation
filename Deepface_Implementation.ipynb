Detect face using Deepface

```python
from deepface import DeepFace # library to find age, gender, emotion and race of any face in a imagem

# detect face
face = DeepFace.detectFace(img_path = img_path, detector_backend = "opencv")
```

```python
# Reduce the dimension of the face array
import numpy as np

if isinstance(face, (list, tuple, np.ndarray)) and len(face) > 0:
  face = face[0] # reducing face from (1, 224, 224, 3) to (224, 224, 3)
```

```python
plt.imshow(face)
```

Analyse Face

```python
obj = DeepFace.analyze(img_path = img_path)
```

```python
obj = obj[0]
```

```python
age = obj['age']
gender = obj['gender']
race = obj['race']
emotion = obj['emotion']
```

```python
print(age, gender, race, emotion)
```

```python
# plotting emotions in a bar plot

import matplotlib.pyplot as plt

# Extract emotion labels and values
emotion_labels = list(emotion.keys())
emotion_values = list(emotion.values())

# Create a bar plot
plt.bar(emotion_labels, emotion_values)
plt.xlabel("Emotions")
plt.ylabel("Probability")
plt.title("Emotion Analysis")
plt.show()
```

Different Backend Function

```python
backends = ["opencv", "ssd", "dlib", "mtcnn", "retinaface", "mediapipe"] # they have different features use them accordingy
```

```python
plt.figure(figsize=(15, 10))  # Adjust the figure size as needed

for i, backend in enumerate(backends):
  try:
    face = DeepFace.detectFace(img_path=img_path, detector_backend=backend)
    if isinstance(face, (list, tuple, np.ndarray)) and len(face) > 0:
      face = face[0]  # reducing face from (1, 224, 224, 3) to (224, 224, 3)
      plt.subplot(2, 3, i + 1)
      plt.imshow(face)
      plt.title(f"Backend: {backend}")
  except Exception as e:
    print(f"Error with backend {backend}: {e}")

plt.tight_layout()
plt.show()
```
